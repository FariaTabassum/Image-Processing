```python

import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
im = cv2.imread('flower.jpg',0)
[M,N]=(im.shape)
T1=125
T2=200
for i in range(M):
    for j in range(N):
        if(im[i,j]>=T1 and im[i,j]<=T2):
            im[i,j]=1
        else:
            im[i,j]=0
plt.imshow(im)
```
