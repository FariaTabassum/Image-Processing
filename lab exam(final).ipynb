```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
im= cv2.imread('flower.jpg',0)
im
```

### Loading the image

```python
plt.imshow(im)
```

### a)Subtracting a constant

```python
[M,N]=(im.shape)
for i in range (M):
    for j in range (N):
        im[i,j] = im[i,j] - 30
        
plt.imshow(im)
```

### b)Single Threshold

```python
[M,N]=(im.shape)
T=125
for i in range (M):
    for j in range(N):
        if im[i,j]<T:
            im[i,j]=0
        else:
            im[i,j]=1
plt.imshow(im)
plt.title("Single Thresholding")
```
