# Temporal Processing


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('circle.jpg',0)
plt.imshow(img)
```

```python
img1 = cv2.imread('sqr.jpg',0)
plt.imshow(img1)
```

## Adding

```python
add_img=img+img1
plt.imshow(add_img)
```

## Subtracting

```python
sub_img = img-img1
plt.imshow(sub_img)
plt.title('Subtracting')
```

## Multiplying

```python
mul_img=img*img1
plt.imshow((mul_img).astype(np.uint8))
plt.title('Multiplying')
```

## Signal Averaging

```python
avg_img=(img+img1)/2
plt.imshow((avg_img).astype(np.uint8))
plt.title('Signal Averaging')
```

## AND

```python
and_img = cv2.bitwise_and(img,img1)
plt.imshow((and_img).astype(np.uint8))
plt.title('AND')
```

## OR

```python
or_img = cv2.bitwise_or(img,img1)
plt.imshow((or_img).astype(np.uint8))
plt.title('OR')
```

## XOR

```python
xor_img = cv2.bitwise_xor(img,img1)
plt.imshow((xor_img).astype(np.uint8))
plt.title('XOR')
```

# Geometric Processing

## Clipping

```python
img2 = cv2.imread('flower.jpg',0)
plt.imshow(img2)
```

```python
y = 60
x= 60
h= 1000
w= 1000
crop_img=img2[x:w,y:h]
plt.imshow(crop_img)
plt.title('Clipping')
```

## Rotating

```python
#rotate 90 clockwise
img_rotate_90_clockwise = cv2.rotate(img2,cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img_rotate_90_clockwise)
plt.title('rotate 90 clockwise')
```

```python
#rotate 90 counter clockwise
img_rotate_90_counterclockwise = cv2.rotate(img2,cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(img_rotate_90_counterclockwise)
plt.title('rotate 90 counterclockwise')
```

```python
#rotate 180
img_rotate_180 = cv2.rotate(img2,cv2.ROTATE_180)
plt.imshow(img_rotate_180)
plt.title('rotate 180 degree')
```

## Flipping

```python
#upsidedown
img_flip_ud = cv2.flip(img2,0)
plt.imshow(img_flip_ud)
plt.title('Flipping on horizontal axis')
```

```python
#leftright and updown
img_flip_up_lr = cv2.flip(img2,-1)
plt.imshow(img_flip_up_lr)
plt.title('Flipping on vertical and horizontal axis')
```

```python
#leftright
img_flip_lr = cv2.flip(img2,1)
plt.imshow(img_flip_lr)
plt.title('Flipping on vertical axis')
```

## Scaling

```python
#zooming
height, width = img2.shape[:2]
zoom_img = cv2.resize(img2,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)
plt.imshow(zoom_img)
plt.title('Zooming')
```

```python
#shrinking
shrink_img = cv2.resize(img2,None,fx=0.2, fy=0.2, interpolation = cv2.INTER_AREA)
plt.imshow(shrink_img)
plt.title('Shrinking')
```
