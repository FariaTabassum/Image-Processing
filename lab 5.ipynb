# Histogram processing

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
flower= cv2.imread('flower.jpg',0)
plt.imshow(flower)
print(flower.shape)
```

```python
flower.ravel()
plt.hist(flower.ravel(),256,[0,256]);
plt.show()
a=np.asarray(flower)
print(a)
```

# Histogram Equalization

```python
equ = cv2.equalizeHist(flower)
out_img = np.hstack((flower,equ))
cv2.imwrite('Equa_image.jpg',out_img)
plt.imshow(out_img)
```
