# Thresholding

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('Rose.jpg',0)
img = cv2.medianBlur(img,5)

ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)

titles = ['Original Image', 'Global Thresholding (v = 85)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]

for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```

# Single thresholding

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('Rose.jpg',0)
[M,N]=(img.shape)
T=60
for i in range (M):
    for j in range(N):
        if img[i,j]<T:
            img[i,j]=0
        else:
            img[i,j]=1
plt.imshow(img)
```

# Dual thresholding

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('Rose.jpg',0)
[M,N]=(img.shape)
T1=15
T2=80
for i in range (M):
    for j in range(N):
        if (img[i,j]>=T1 and img[i,j]<=T2):
            img[i,j]=1
        else:
            img[i,j]=0
plt.imshow(img)
```

# Gray Level Slicing 1st approach


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('Rose.jpg',0)
[M,N]=(img.shape)
A=10
B=60
for i in range (M):
    for j in range(N):
        if (img[i,j]>=A and img[i,j]<=B):
            img[i,j] =255 #l-1
        else:
            img[i,j]=0
        
plt.imshow(img)
```
