```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
Rose = cv2.imread('Rose.jpg',0)
plt.imshow(Rose)
print(Rose.shape)
```

# 1. Histogram processing

```python
import cv2
import numpy as np
Rose.ravel()
plt.hist(Rose.ravel(),230,[0,230]);
plt.show()
a=np.asarray(Rose)
print(a)
```

# 2. Histogram Equalization

```python
equ = cv2.equalizeHist(Rose)
out_Rose = np.hstack((Rose,equ))
cv2.imwrite('Equa_Rose.jpg',out_Rose)
plt.imshow(out_Rose)
```
