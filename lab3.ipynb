# Monadic Operations

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
img = cv2.imread('flower.jpg',0)
plt.imshow(img)
```

```python
#Adding a constant
[M,N]=(img.shape)
for i in range (M):
    for j in range (N):
        img[i,j] = img[i,j] + 30
        
plt.imshow(img)
```

```python
#Subtracting a constant
[M,N]=(img.shape)
for i in range (M):
    for j in range (N):
        img[i,j] = img[i,j] - 30
        
plt.imshow(img)
```

```python
#Negative
[M,N]=(img.shape)
for i in range (M):
    for j in range (N):
        img[i,j] = 300 - img[i,j]
        
plt.imshow(img)
```

```python
#Multiplying into constant
[M,N]=(img.shape)
for i in range (M):
    for j in range (N):
        img[i,j] = img[i,j] * 30
        
plt.imshow(img)
```

```python
#Multiplying by a constant
[M,N]=(img.shape)
for i in range (M):
    for j in range (N):
        img[i,j] = 30 * img[i,j]
        
plt.imshow(img)
```

```python
# Divide by a constant
[M,N]=(img.shape)
for i in range (M):
    for j in range (N):
        img[i,j] = img[i,j]/ 30
        
plt.imshow(img)
```

```python
#AND
and_img = cv2.bitwise_and(img,30)
plt.imshow(and_img)
```

```python
#OR
or_img = cv2.bitwise_or(img,30)
plt.imshow(or_img)
```

```python
#XOR
xor_img = cv2.bitwise_xor(img,30)
plt.imshow(xor_img)
```
