```python
import cv2
import numpy as np
from IPython.display import Image
from matplotlib import pyplot as plt
from matplotlib import pyplot as plt1
from matplotlib import pyplot as plt2
```

```python
img = cv2.imread("flower.jpg",0)
plt.imshow(img)
```

```python
r1=80
r2=200
s1=30
s2=210
```

```python
al=s1/r1
bt=(s2-s1)/(r2-r1)
gm=(255-s2)/(255-r2)
c1=s1-bt*r1
c2=s2-gm*r2
d=img
for i in range(len(img)):
    for j in range(len(img[0])):
        if(img[i][j] < r1):
            d[i][j] = al*img[i][j]
        elif( r1 <= img[i][j] < r2 ):
            d[i][j] = bt*img[i][j] + c1
        else:
            d[i][j] = gm*img[i][j] + c2
cv2.imwrite('flower_contrast.jpg',d)
Image('flower.jpg')
Image('flower_contrast.jpg')
```
