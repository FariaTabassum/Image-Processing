```python
img1 = cv2.imread('flower.jpg',0)
[M,N] = img1.shape
```

```python
#contrast stretching method 1
t1=80
t2=200

for i in range (M):
    for j in range (N):
        if (img1[i,j]>=t1) & (img1[i,j]<=t2):
            img1[i,j]=(227 * img1[i,j] -5040)/47
        else:
            pass
plt.imshow(img1)
```

```python
#contrast stretching method 2
t1=30
t2=210

for i in range (M):
    for j in range (N):
        if (img1[i,j]<t1):
            img1[i,j]= 0
        elif (img1[i,j]>=t1) & (img1[i,j]<=t2):
            img1[i,j]=(255 * img1[i,j] -22950)/48
        elif (img1[i,j]>t2):
            img1[i,j]=255
plt.imshow(img1)
```

```python
#log transformation
img = cv2.imread('flower.jpg',1)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
c = 255 / np.log(1 + np.max(img))
print(c)
log_image = c * (np.log(img + 1))
   
log_image = np.array(log_image, dtype = np.uint8)
   
plt.imshow(img)
plt.show()
plt.imshow(log_image)
plt.show()
```

```python
#power law transformation
img = cv2.imread('flower.jpg',0)
[M,N] = img.shape
c = 255 / np.log(1 + np.max(img))
for i in range (M):
    for j in range (N):
        img[i,j] = c * pow((img[i,j]),(1/0.9))
plt.imshow(img)
```

```python
img = cv2.imread('flower.jpg',0)
[M,N] = img.shape
c = 255 / np.log(1 + np.max(img))
for i in range (M):
    for j in range (N):
        img[i,j] = c * pow((img[i,j]),(0.78))
plt.imshow(img)
```
